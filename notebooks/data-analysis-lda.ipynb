# Data Analysis for Twitter and E-participation Platforms

This notebook shows the analysis process for twitter and e-participation data used in our paper.
We propose a mapping between the five steps of policy-making as theorized by Howlett and al. (2009) and text analysis methods enabling to extract valuable knowledge from text written citizens for each stage of the policy making process. The steps are the following :

- **Agenda-Setting** : Framing the problem and exploring the needs for a policy,
- **Formulation** : Developing policy options and informing them,
- **Decision** : Selection of the option among a range of options,
- **Implementation** : enactment of the passed policies through legislation, regulation, planning,
- **Monitoring** : Evaluation of the effects of passed policy

```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import glob
import unidecode
from spacy.lang.fr.stop_words import STOP_WORDS as fr_stop
import random
import nltk

stop_words_fr = list(fr_stop)+['qu', 'quelqu']

# Fixing Random
np.random.seed(5000)
random.seed(5000)
```

```python
# Loading the twitter data
DATA_FOLDER = "../data"

path = "{}/{}".format(DATA_FOLDER, "all_tweets.csv")
tweets = pd.read_csv(path)

# Loading participations data
path = "{}/{}".format(DATA_FOLDER, "reinventons-liege.csv")
reinventons = pd.read_csv(path, delimiter=";", header=0,  names=["cat","title", "pos", "neg", "desc"])

# Cleaning the artifact in the participation data
clean = []
for i, row in reinventons.iterrows():
    text = row["desc"].replace('\r', ' ').replace('\n', ' ').replace(";", "")
    text = unidecode.unidecode(text).lower()
    clean.append(text)
reinventons["desc"] = clean

# LOG
print("Number of tweets {}".format(len(tweets)))
print("Number of citizens ideas (Réinventons Liège) {}".format(len(reinventons)))
```

```python
from nltk.stem.snowball import FrenchStemmer

stemmer = FrenchStemmer()

# Retrieving the text data from the two datasets
text_data = [x[1]["cleaned_text"] for x in tweets.iterrows()]
text_data += [x[1]["desc"] for x in reinventons.iterrows()]

stemmed_data = []

# Stemming the text data
for text in text_data:
    stemmed_data.append(stemmer.stem(text))
        
```

# Agenda Setting

During this phase, the decision makers will set up their agenda and find interesting issues to address during their term office.
A clustering is created in order to explore the text data and find relevant cluster for further analysis

The clustering is made with the **Latent Dirichelet Allocation** algorithm from [scikit-learn](https://scikit-learn.org/stable/) and the texts are preprocessed using **TF-IDF**

The visualization of the clustering is made using the **T-SNE** algorithm

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.manifold import TSNE 

# Compute TF-IDF for each text 
vectorizer = TfidfVectorizer(stop_words=stop_words_fr)
X = vectorizer.fit_transform(text_data)

# Computing various k-means
nbr_clusters_to_keep = 30
score = []
clusters = None

for nbr_cluster in range(20, 50):
    classifier = LatentDirichletAllocation(nbr_cluster)
    
    if nbr_cluster == nbr_clusters_to_keep:
        clusters = classifier.fit_transform(X)
        score.append(classifier.score(X))
    else:
        classifier.fit(X)
        score.append(classifier.score(X))

    
cluster_idx = []

for elem in clusters:
    cluster_idx.append(np.argmax(elem))

# TSNE dimension reduction + display
X_embedded = TSNE(n_components=2).fit_transform(clusters)

data = pd.DataFrame(X_embedded, columns =['tsne1', 'tsne2'])
data["clusters"] = cluster_idx
data["val"] = text_data
data["stemmed_val"] = stemmed_data
```

```python
# Display kmeans scores 
plot_data = {
    "score": score,
    "nbr_cluster": range(20, 50)
}

sns.lineplot(data=plot_data, x="nbr_cluster", y="score")
```

```python
# Show plot
plt.figure(figsize=(15,10))
sns.scatterplot(x = "tsne1", y="tsne2", hue="clusters", data=data, palette=sns.color_palette("hls", nbr_clusters_to_keep))
```

```python
# Exploring clusters :
import random 

clusters_to_show = [3, 6, 7, 8, 10, 23, 17, 28]
percentage_to_show = 0.15


for i in clusters_to_show:
    print("\n\n\n=======================================================")
    print("CLUSTER {}".format(i))
    print("=======================================================\n\n\n")
    
    for j, row in data.iterrows():
        if row["clusters"] == i and random.random() < percentage_to_show:
            
            print(row["val"] + "\n")
            print("------\n")
            
```

# Formulation Phase

The policy makers identified an interesting cluster in the previous phase. They could isolate it and perform another clustering in order to identify interesting sub-subject or various opinions in the cluster. In this example we keep the cluster 3 with text focusing on alternative mobility (bike principally).

The new division obtained divide the text in two category. Category 0 contains a lot of text talking about the mobility near the river "Meuse" while the other one tend to focus more on parc and nature. 

```python
from sklearn.cluster import KMeans

to_keep = 3

keeped = data[data["clusters"] == to_keep]
print("Number of texts in the cluster : {}".format(len(keeped)))

# KMeans :
nbr_cluster = 2

# Computing vector for texts
vectorizer = TfidfVectorizer(stop_words=stop_words_fr)
X = vectorizer.fit_transform(keeped["val"].to_list())

# KMeans
f_kmeans = KMeans(n_clusters=nbr_cluster).fit_predict(X)

# TSNE
X_embedded = TSNE(n_components=2).fit_transform(X)

f_data = pd.DataFrame(X_embedded, columns =['tsne1', 'tsne2'])
f_data["clusters"] = f_kmeans
f_data["val"] = keeped["val"].to_list()
```

```python
# Plotting

plt.figure(figsize=(15,10))
sns.scatterplot(x = "tsne1", y="tsne2", hue="clusters", data=f_data, palette=sns.color_palette("hls", nbr_cluster))
```

```python
for i in range(nbr_cluster):
    print("\n\n\n=======================================================")
    print("CLUSTER {}".format(i))
    print("=======================================================\n\n\n")
    
    for j, row in f_data.iterrows():
        if row["clusters"] == i:
            print(row["val"] + "\n")
            print("------\n")
```

# Monitoring

Once a policy have been developped, the policy maker may wish to investigate if its policy have impact on the citizen. They could monitore social media in order to detect a change in the appreciation of the citizens.

We perform the analysis on the cluster 15 grouping text about Covid-19. The resulting visualization allow the policy maker to monitor the sentiment vehiculed in tweets and when

```python
# Preprocessing

from textblob import Blobber
from textblob_fr import PatternTagger, PatternAnalyzer

to_monitor = 15
monitor = data[data["clusters"] == to_monitor]

# Perform sentiment analysis

def find_timestamp(text):
    for j, elem in tweets.iterrows():
        if row["val"] == elem["cleaned_text"]:
            return elem["created_at"]
        
    return ""
            
    

sentiments = []
timestamp = []

for i, row in monitor.iterrows():
    tb = Blobber(pos_tagger=PatternTagger(), analyzer=PatternAnalyzer())
    blob = tb(row["val"])
    if blob.sentiment[0] < 0:
        sentiments.append(-1)
    elif blob.sentiment[0] > 0:
        sentiments.append(1)
    else :
        sentiments.append(0)
    
    timestamp.append(find_timestamp(row["val"]))
            
monitor["sentiment"] = sentiments
monitor["created_at"] = timestamp

monitor = monitor[monitor["created_at"] != ""]
```

```python
from matplotlib.colors import ListedColormap

cmap = ListedColormap(["red", "blue", "green"])

to_plot = monitor[["created_at", "sentiment"]].copy()

to_plot['created_at'] = pd.to_datetime(to_plot['created_at'])
to_plot.index = to_plot['created_at'] 

by_day_sentiment = to_plot.groupby([pd.Grouper(key='created_at', freq='D'), 'sentiment']).size().unstack('sentiment')

by_day_sentiment.plot(kind="bar", figsize=(20,3), color=["red", "blue", "green"])
```
