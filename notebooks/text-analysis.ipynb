# Chargement + nettoyage des données 

Toutes les données sont chargées, mais seuls les données des descriptions sont nettoyées, Les retour à la ligne, tabulation et ; sont retirés. Les caractère accentués ont aussi du être enlevé car ceux-ci rendent le travail difficiles pour certains algorithmes (maudits anglais!)

```python
# loading the data + basic analysis
#df = pd.read_csv("../data/dataV2.csv", delimiter=";", header=0,  names=["num","titre", "cat","quartier", "auteur", "pos", "neg", "desc", "desc2", "etc"])
df = pd.read_csv("../data/reinventons-liege.csv", delimiter=";", header=0,  names=["cat","title", "pos", "neg", "desc"])

# Clean desc
clean = []
for i, row in df.iterrows():
    text = row["desc"].replace('\r', ' ').replace('\n', ' ').replace(";", "")
    text = unidecode.unidecode(text).lower()
    clean.append(text)
print(len(df))
df["desc"] = clean
```

# Texte clustering

L'objectif des prochaines manoeuvre est de visualiser les textes en 2 dimensions. L'intuition est que l'on pourrait tenter de diviser l'ensembles des idées en "cluster" de texte similaire. 

**Question : Qu'est-ce qu'un texte similaire**
Il y a plusieurs façon de calculer la "similarité" de textes. La plus répendue est de calculer la distance consinus. Malheureusement cette méthode n'est pas adaptées dans le cas ou l'on compare plus de 2 textes.

Ce qu'il faut faire c'est calculer un "descripteur" de texte qui serait un vecteur de nombres. Choses que l'on sait manipuler avec des algorithmes classiques. Cela peut ce faire grâce à l'algo TF-IDF : https://fr.wikipedia.org/wiki/TF-IDF

Ainsi on peut utiliser des algo de réduction de dimensions et de clustering classique. A savoir TF-IDF et K-MEANS.

```python
# Trying with simple tf-idf

# Cleaning
def preprocessing(line):
    line = line.lower()
    line = re.sub(r"[{}]".format(string.punctuation), " ", line)
    return line

documents = df["desc"].tolist()
vectorizer = TfidfVectorizer(max_df=0.8, max_features=200000, min_df=0.2, stop_words=list(fr_stop), use_idf=True, preprocessor=preprocessing)
X = vectorizer.fit_transform(documents)

X_embedded = TSNE(n_components=2).fit_transform(X)
kmeans = KMeans(n_clusters=6).fit_predict(X)
```

# Visu

Cette visu va nous permettre de comparer le clustering calculé par similarité de texte aux catégories qui ont été définies par les agents communaux. Evidemment les données ne correspondent pas au clustering et ces méthodes d'IA peuvent permettre d'obtenir un autre clustering qui pourrait être pertinent.

**Probleme : Le choix du nom pour les clusters créés a été fait naivement en prenant le mot le plus représentatif du corpus selon tf-idf. Néanmoins ce n'est pas toujours le meilleur il faudrait améliorer cela**

```python
def get_word_mapping(df):
    mapping = {}
    
    for name, group in df.groupby(['kmeans']):
        tfidf = TfidfVectorizer(stop_words=list(fr_stop))
        word_count_vector=tfidf.fit_transform(group["desc"])
        feature_array = np.array(tfidf.get_feature_names())
        tfidf_sorting = np.argsort(word_count_vector.toarray()).flatten()[::-1]
        
        if name == 2:
            mapping[name] = "disparitions"
        else:
            mapping[name] = feature_array[tfidf_sorting][2]
        
    return mapping


tsne = pd.DataFrame(X_embedded, columns =['tsne1', 'tsne2'])

df["tsne1"] = tsne["tsne1"]
df["tsne2"] = tsne["tsne2"]
df["kmeans"] = kmeans
mapping = get_word_mapping(df)

df["kmeans_label"] = [mapping[x] for x in df["kmeans"]]

plt.figure(figsize=(15,10))
sns.scatterplot(x = "tsne1", y="tsne2", hue="cat", data=df)
plt.savefig("classification-agent.jpg")

plt.figure(figsize=(15,10))
sns.scatterplot(x = "tsne1", y="tsne2", hue="kmeans_label", data=df)
plt.savefig("classification-pc.jpg")
```
