# Data Analysis for Twitter and E-participation Platforms

This notebooks shows the analysis process for twitter and e-participation data used in our paper.
We propose a mapping between the fives steps of policy-making as theorized by Howlett and al. (2009) and text analysis methods enabling to extract valuable knowledge from text written by the citizens for each stage of the policy making process. The steps are the following :

- **Agenda-Setting** : Framing the problem and exploring the needs for a policy,
- **Formulation** : Developing policy options and informing them,
- **Decision** : Selection of the option among a range of options,
- **Implementation** : enactment of the passed policies through legislation, regulation, planning,
- **Monitoring** : Evaluation of the effects of passed policy

```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import glob
import unidecode
from spacy.lang.fr.stop_words import STOP_WORDS as fr_stop

stop_words_fr = list(fr_stop)+['qu', 'quelqu']
```

```python
# Loading the twitter data
DATA_FOLDER = "../data"

path = "{}/{}".format(DATA_FOLDER, "all_tweets.csv")
tweets = pd.read_csv(path)

# Loading participations data
path = "{}/{}".format(DATA_FOLDER, "reinventons-liege.csv")
reinventons = pd.read_csv(path, delimiter=";", header=0,  names=["cat","title", "pos", "neg", "desc"])

# Cleaning the artifact in the participation data
clean = []
for i, row in reinventons.iterrows():
    text = row["desc"].replace('\r', ' ').replace('\n', ' ').replace(";", "")
    text = unidecode.unidecode(text).lower()
    clean.append(text)
reinventons["desc"] = clean

# LOG
print("Number of tweets {}".format(len(tweets)))
print("Number of citizens ideas (Réinventons Liège) {}".format(len(reinventons)))
```

# Agenda Setting

During this phase, the decision makers will set up their agenda and find interesting issue to address during their term office.
A clustering is created in order to explore the text data and find relevant cluster for further analysis

The clustering is made with the **K-MEANS** algorithm from [scikit-learn](https://scikit-learn.org/stable/) and the texts are preprocessed using **TF-IDF**

The visualization of the clustering is made using the **T-SNE** algorithm

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.manifold import TSNE 

# Retrieving the text data from the two datasets

text_data = [x[1]["cleaned_text"] for x in tweets.iterrows()]
text_data += [x[1]["desc"] for x in reinventons.iterrows()]

# Compute TF-IDF for each text 
vectorizer = TfidfVectorizer(stop_words=stop_words_fr)
X = vectorizer.fit_transform(text_data)

# Computing K-MEANS
nbr_cluster = 30
kmeans = KMeans(n_clusters=nbr_cluster).fit_predict(X)

# TSNE dimension reduction + display
X_embedded = TSNE(n_components=2).fit_transform(X)

data = pd.DataFrame(X_embedded, columns =['tsne1', 'tsne2'])
data["kmeans"] = kmeans
data["val"] = text_data
```

```python
# Show plot
plt.figure(figsize=(15,10))
sns.scatterplot(x = "tsne1", y="tsne2", hue="kmeans", data=data, palette=sns.color_palette("hls", nbr_cluster))
```

```python
# Exploring clusters :
import random 

clusters_to_show = [3, 7, 8, 9, 10, 15, 17, 29]
percentage_to_show = 0.15


for i in clusters_to_show:
    print("\n\n\n=======================================================")
    print("CLUSTER {}".format(i))
    print("=======================================================\n\n\n")
    
    for j, row in data.iterrows():
        if row["kmeans"] == i and random.random() < percentage_to_show:
            
            print(row["val"] + "\n")
            print("------\n")
            
```

# Monitoring

Once a policy have been developped, the policy maker may wish to investigate if its policy have impact on the citizen. They could monitore social media in order to detect a change in the appreciation of the citizens.

We perform the analysis on the cluster 15 grouping text about Covid-19. The resulting visualization allow the policy maker to monitor the sentiment vehiculed in tweets and when

```python
# Preprocessing

from textblob import Blobber
from textblob_fr import PatternTagger, PatternAnalyzer

to_monitor = 15
monitor = data[data["kmeans"] == to_monitor]

# Perform sentiment analysis

def find_timestamp(text):
    for j, elem in tweets.iterrows():
        if row["val"] == elem["cleaned_text"]:
            return elem["created_at"]
        
    return ""
            
    

sentiments = []
timestamp = []

for i, row in monitor.iterrows():
    tb = Blobber(pos_tagger=PatternTagger(), analyzer=PatternAnalyzer())
    blob = tb(row["val"])
    if blob.sentiment[0] < 0:
        sentiments.append(-1)
    elif blob.sentiment[0] > 0:
        sentiments.append(1)
    else :
        sentiments.append(0)
    
    timestamp.append(find_timestamp(row["val"]))
            
monitor["sentiment"] = sentiments
monitor["created_at"] = timestamp

monitor = monitor[monitor["created_at"] != ""]
```

```python
from matplotlib.colors import ListedColormap

cmap = ListedColormap(["red", "blue", "green"])

to_plot = monitor[["created_at", "sentiment"]].copy()

to_plot['created_at'] = pd.to_datetime(to_plot['created_at'])
to_plot.index = to_plot['created_at'] 

by_day_sentiment = to_plot.groupby([pd.Grouper(key='created_at', freq='D'), 'sentiment']).size().unstack('sentiment')

by_day_sentiment.plot(kind="bar", figsize=(20,3), color=["red", "blue", "green"])
```
